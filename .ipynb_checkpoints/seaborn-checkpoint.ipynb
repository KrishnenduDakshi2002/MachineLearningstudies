```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
## here tip->> dependent feature 
## total_bill, sex, smoker, day,time, size --> independent features
# because tip will depend on other features
```

# Correlation with Heatmap

```python
# A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix(table) between two discrete
# dimensions or event types, It is very important in Feature selection
```

```python
df.corr()  ## correlation is only for int or float (not for objects)
```

```python
sns.heatmap(df.corr())
```

# Joint plot

```python
sns.jointplot(x='tip',y='total_bill',data =df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data =df,kind='reg')
```

# Pair plot

```python
## When we have more than 2 independent features we use pair plot
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

# Dist plot

```python
sns.displot(df['tip'],kde=True)
```

# Categorical plot
    boxplot
    violinplot
    countplot
    bar plot

```python
## count plot

sns.countplot('sex',data=df)
```

```python
## bar plot

sns.barplot(x='total_bill',y='smoker',data=df)
```

```python
## box plot

sns.boxplot('smoker','total_bill',data=df)
```

```python
# categorize my data based on some other categories

sns.boxplot(x='total_bill',y='day',hue='smoker',data = df)
```

```python
# violin plot

# it helps us to see both the distribution of data in terms of kernal density estimation an dthe box plot

sns.violinplot(x='total_bill',y='day',hue='smoker',data=df,palette = 'rainbow')
```
