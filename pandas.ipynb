```python
import numpy as np
import pandas as pd
```

```python
# playing with Dataframe

df = pd.DataFrame(np.arange(0,20).reshape(5,4),index =[ 'Row1','Row2','Row3','Row4','Row5'],columns=['Column1','Column2','Column3','Column4'])
```

```python
df.head()
```

```python
df.to_csv('test.csv') # export df to csv file
```

```python
## Accessing the elements
## 1. .loc  2. .iloc

df.loc['Row2']  ## getting a series
```

```python
df['Column2']
```

```python
df[['Column2','Column4']] ## getting multiple columns
```

```python
type(df.loc['Row3'])  ## series can be an entire one row or an entire column
# but dataframe must have altest more than one row or one column
```

```python
df.iloc[2:,2:] ## row index , column index
```

```python
type(df.iloc[2:,2:]) ## this gives dataframe
```

```python
type(df.iloc[2:,0]) # now we are getting series
```

```python
## convert dataframe into array
df.iloc[:,:].values
```

```python
df.iloc[:,:].values.reshape(2,10)
```

```python
## how to check null condition
df.isnull()
# or 
df.isnull().sum()
```

```python
df['Column2'].value_counts()
```

```python
df['Column2'].unique()
```
