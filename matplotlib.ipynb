```python
import matplotlib.pyplot as plt
%matplotlib inline   
##if we don't write this line we have to write "plt.show()" every time 
```

```python
import numpy as np
```

```python
## simple examples 
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
## plotting using matplotlib

## plt scatter

plt.scatter(x,y,c='b')

#titles
plt.title("Graph")

# title X and Y axis
plt.xlabel('x axis')
plt.ylabel('y axis')

#saving plot as png
plt.savefig("plot.png")
```

```python
y = x*x
```

```python
plt.plot(x,y,'ro',linestyle= 'dashed',linewidth=2)


#titles
plt.title("Graph")

# title X and Y axis
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
#creating subplot

plt.subplot(2,2,1)
plt.plot(x,y,'r')


plt.subplot(2,2,2)
plt.plot(x,y,'o')


plt.subplot(2,2,3)
plt.plot(x,y,'b')


plt.subplot(2,2,4)
plt.plot(x,y,'g')
```

# Box plot using matplotlib

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
```

```python
plt.boxplot(data,vert=True,patch_artist=True)
plt.show()
```

# Pie charts

```python
# data to plot
labels = 'python','c++', 'Ruby', 'Java'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightblue']
explode = (0.1,0,0,0) # explode 1st slice

#plot
plt.pie(sizes,explode = explode , labels = labels, colors = colors, autopct= '%1.1f%%', shadow = True)
plt.axis("equal")
plt.show()
```
